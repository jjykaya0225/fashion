```python
merge_data = pd.merge(order_data, customer_data, on = "고객번호", how = "left")
```

```python
merge_data.head()
```

```python
merge_data["주문일자"] = merge_data["주문일시"].str.slice(start=0, stop=10)
```

```python
merge_data["주문일자"] = pd.to_datetime(merge_data["주문일자"], format = "%Y-%m-%d")
```

```python
merge_data["가입일자"] = merge_data["가입일자"].str.slice(start=0, stop=10)
merge_data["가입일자"] = pd.to_datetime(merge_data["가입일자"], format = "%Y-%m-%d")
```

```python
merge_data
```

```python
# 상품번호 컬럼에서 브랜드명/시즌/카테고리 추출하기
merge_data["브랜드명"] = merge_data["상품번호"].str.slice(3,5)
merge_data["시즌"] = merge_data["상품번호"].str.slice(5,9)
merge_data["카테고리"] = merge_data["상품번호"].str.slice(9,12)
```

```python
merge_data.head()
```

```python
#범주별 분포
brand_type = merge_data["브랜드명"].value_counts()
```

```python
brand_type
```

```python
%matplotlib inline
plt.bar(brand_type.index, brand_type.values, width=0.8, align='edge')
plt.show()
```

```python
category_type = merge_data["카테고리"].value_counts()
%matplotlib inline
plt.bar(category_type.index, category_type.values, width=0.8, align='edge')
plt.show()
```

```python
#연속형/이산형 -> 데이터 범주화
bins = list(range(10,100,10))
```

```python
bins
```

```python
bins_label = [str(x) + "대" for x in bins]
```

```python
bins_label
```

```python
merge_data["나이대"] = pd.cut(merge_data["나이"], bins, right = False, labels = bins_label[:-1])
```

```python
merge_data
```

```python
merge_data.loc[(merge_data["정상가"]>0)&(merge_data["정상가"]<=100000), "가격대"] = "0만원~10만원"
merge_data.loc[(merge_data["정상가"]>100000)&(merge_data["정상가"]<=200000), "가격대"] = "10만원~20만원"
merge_data.loc[(merge_data["정상가"]>200000)&(merge_data["정상가"]<=500000), "가격대"] = "20만원~50만원"
merge_data.loc[(merge_data["정상가"]>500000)&(merge_data["정상가"]<=1000000), "가격대"] = "50만원~100만원"
merge_data.loc[(merge_data["정상가"]>1000000), "가격대"] = "100만원 ~"
```

```python
merge_data
```

```python
merge_data["가격대"].value_counts()
```

```python
#groupby
merge_data.groupby("브랜드명").agg({"정상가":"mean", "판매가":"max", "고객번호":"nunique", "주문번호":"nunique"})
```

```python
merge_data.groupby(["브랜드명", "카테고리"]).agg({"정상가":"mean", "판매가":"max", "고객번호":"nunique", "주문번호":"nunique"})
```

```python
ap_data = merge_data[merge_data["브랜드명"]=="AP"]
plt.title("AP 브랜드의 정상가 분포")
plt.hist("정상가", bins = 30, range = (10000, 1000000), color = "purple", data = ap_data)
plt.show()
```
